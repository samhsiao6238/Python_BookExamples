```python
"""
金融歷史資料套件應用
yahoo finance ：金融商品資料取得
安裝套件
# pip install yfinance 
"""

# 載入套件
import yfinance as yf

# 商品名稱 GSPC，標準普爾 S&P 500 的代碼
# 用來區分普通股票代號與特定類型的金融指數或金融商品的識別符號
prod = "^GSPC"

# 取得歷史資料 yf.download()
data1 = yf.download(
    prod,
    start="2024-01-01",
    end="2024-03-31"
)

# 取得全部資料
data2 = yf.download(
    prod,
    period="max"
)

data1, data2
```
